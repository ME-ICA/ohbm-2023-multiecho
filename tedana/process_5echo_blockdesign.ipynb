```python
%matplotlib inline
import os
import os.path as op
from glob import glob
import webbrowser

from tedana.workflows import tedana_workflow, ica_reclassify_workflow
```

# Set up data download and tedana output directories

```python
%%time
dset_dir = 'five-echo/'
out_dir_kundu = 'tedana_results_kundu_five-echo'
out_dir_minimal = 'tedana_results_minimal_five-echo'
out_dir_reclass = 'ica_reclassify_five-echo'
wd = os.getcwd()

if not op.isdir(dset_dir):
    os.mkdir(dset_dir)

if not op.isdir(out_dir_kundu):
    os.mkdir(out_dir_kundu)

if not op.isdir(out_dir_minimal):
    os.mkdir(out_dir_minimal)

if not op.isdir(out_dir_reclass):
    os.mkdir(out_dir_reclass)

os.chdir(dset_dir)
```

## Run workflow using "kundu" component selection process for denoising

```python
%%time
os.chdir(wd)
files = sorted(glob(op.join(dset_dir, 'p06.SBJ01_S09_Task11_e*.sm.nii.gz')))
tes = [15.4, 29.7, 44.0, 58.3, 72.6]
tedana_workflow(files, tes, out_dir=out_dir_kundu, tree='kundu', tedpca='mdl', convention='bids', tedort=False, fixed_seed=42)
```

# Run using the same ICA mixing matrix and the "minimal" component selection process

```python
%%time
os.chdir(wd)
files = sorted(glob(op.join(dset_dir, 'p06.SBJ01_S09_Task11_e*.sm.nii.gz')))
tes = [15.4, 29.7, 44.0, 58.3, 72.6]
print(op.join(out_dir_kundu, 'desc-ICA_mixing.tsv'))
tedana_workflow(files, tes, out_dir=out_dir_minimal, tree='minimal', mixm=op.join(out_dir_kundu, 'desc-ICA_mixing.tsv'), convention='bids', tedort=False)
```

## Open tedana report using "minimal" component selection process in web browser

Note that component 30 was rejected with kundu and accepted with minimal. It is the main visual cortex response to a block design checkerboard and it clearly should have been accepted
To get a better sense of what was different in kundu [desc-ICA_status_table.tsv](./tedana_results_kundu_five-echo/desc-ICA_status_table.tsv) shows in which step each component's classification changed.
Component 30 was classified as "unclass_highvar" in Node 7 and then "rejected" in node 17. The names of each of these nodes and more explaination is in [desc-ICA_decision_tree.json](./tedana_results_kundu_five-echo/desc-ICA_decision_tree.json). A visual flow chart of the steps is part of the [tedana documentation](https://tedana.readthedocs.io/en/stable/included_decision_trees.html). Specifically in this case, the visual cortex response to the checkerboard is very large and the high variance compared to other components meant that it was rejected based on the threshold in node 17, which rejects some components with high variance and a relatively high rho value (S0 weighting), even if the kappa value (T2* weighting) is also very high.

```python
new = 2 # open in a new tab, if possible

url = str(os.path.abspath(op.join(out_dir_minimal, 'tedana_report.html')))
print(url)
webbrowser.open("file://" + url,new=new)
```
